```python
data_pickle = pd.read_pickle("datos/data_final_encoded.pkl")
df = pd.DataFrame(data_pickle)
df.head()
```

```python
lista_modelos = ["logistic_regression", "tree", "random_forest", "gradient_boosting", "xgboost"]
df_metricas = pd.DataFrame()
soporte = sle.AnalisisModelosClasificacion(df, "Attrition")
for modelo in lista_modelos:
    print(f"------------------ Uilizando {modelo.upper()} ------------------")
    print("----------------------------------------------------------------")
    soporte.ajustar_modelo(modelo)
    df_temp = soporte.calcular_metricas(modelo)
    df_metricas = pd.concat([df_metricas, df_temp], axis=1)
    soporte.plot_matriz_confusion(modelo)
    soporte.importancia_predictores(modelo)
    soporte.hacer_roc_curve(modelo)
    soporte.plot_shap_summary(modelo)
    display(df_metricas)
df_metricas.head().T
```

# Conclusiones

1. **Modelos de Ensamble Sobresalen**:
   - **Random Forest** muestra una `accuracy` de **100%** en el conjunto de entrenamiento, mientras que en el conjunto de prueba su `accuracy` baja a **92.7%**. Los valores de `kappa` son **1.0** para el entrenamiento y **0.85** en la prueba, lo cual indica que el modelo generaliza bien, aunque muestra una ligera caída en la precisión en el conjunto de prueba, sugiriendo un leve **overfitting**.

2. **Potencial Overfitting en Modelos Boosting**:
   - Tanto **Gradient Boosting** como **XGBoost** presentan métricas perfectas en el conjunto de entrenamiento (`accuracy`, `precision`, `recall`, `f1`, `kappa`, `auc` de **1.0**), lo que sugiere un posible **sobreajuste**. En el conjunto de prueba, **Gradient Boosting** obtiene una `accuracy` de **92.3%**, y **XGBoost** de **92.5%**, lo cual es ligeramente menor, pero aún indica un buen rendimiento general.

3. **Árbol de Decisión**:
   - El **Árbol de Decisión** tiene una `accuracy` de **94.6%** en el conjunto de entrenamiento y una de **83.9%** en el conjunto de prueba. La métrica de `kappa` también se reduce de **0.89** a **0.68** entre los conjuntos, indicando un **sobreajuste** significativo, ya que el modelo tiene dificultades para mantener el mismo nivel de rendimiento en el conjunto de prueba.

4. **Regresión Logística**:
   - La **Regresión Logística** presenta una `accuracy` de **80.3%** en el conjunto de entrenamiento y de **80.5%** en el conjunto de prueba. Los valores de `kappa` son **0.61** en el conjunto de entrenamiento y **0.61** en el de prueba, lo cual indica un rendimiento constante, aunque moderado. Este modelo proporciona una buena base, pero no alcanza la precisión de los modelos de ensamble.
